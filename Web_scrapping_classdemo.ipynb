```python
import bs4  #bs version 4
from urllib.request import urlopen  #To pass url,u give urlopen
import pandas as pd #used for data ananlysis and manipulation

#check robots.txt before any web scraping

response = urlopen("https://www.amazon.in/s?k=mobiles&ref=nb_sb_noss_2") #target url

#send request to downlaod the data
html_doc = response.read()
               
#parse the download data
soup =bs4.BeautifulSoup(html_doc,'html.parser')
print(soup.prettify())
  
```

```python
mobile_name_element=soup.find_all("span",class_="a-size-medium a-color-base a-text-normal")
mobile_price_element=soup.find_all("span",class_="a-price-whole")
print("The total items:",len(mobile_name_element))
print("The total price of items available: ",len(mobile_price_element))


for i in range(len(mobile_name_element)):
    mobile_name=mobile_name_element[i].text
    mobile_price=mobile_price_element[i].text
    print(mobile_name,mobile_price)
```

```python
# Create a list to store the data
scraped_data=[]
for i in range(len(mobile_name_element)):
    #Intialize the dictionary
    dict={}
    
    # Scraping the mobile name
    mobile_name_element=soup.find_all("span",class_="a-size-medium a-color-base a-text-normal")
    
    # Scraping the mobile price
    mobile_price_element=soup.find_all("span",class_="a-price-whole")
    
    #Iterating the items one by one
    mobile_name=mobile_name_element[i].text
    mobile_price=mobile_price_element[i].text
    
    # Storing it in the dict
    dict['mobile_name']= mobile_name
    dict['mobile_price']=mobile_price
    
    #Appending scraped data to the list
    scraped_data.append(dict)
    
#Create a dataframe from the list of dictionaries
dataFrame=pd.DataFrame.from_dict(scraped_data)
dataFrame.to_csv('Amazon_Mobile_data.csv',index=False)
```

```python
import bs4
from urllib.request import urlopen
import pandas as pd

response = urlopen("https://www.amazon.sg/s?k=hauwei&rh=n%3A6436149051&ref=nb_sb_noss")
html_doc = response.read()
soup =bs4.BeautifulSoup(html_doc,'html.parser')
print(soup.prettify())

mobile_name_element = soup.find_all("span",class_="a-size-base-plus a-color-base a-text-normal")
mobile_price_element  = soup.find_all("span",class_="a-price-whole")

print("Total mobiles available:",len(mobile_name_element))
print("Total prices available:",len(mobile_price_element))
print()

scrape_data=[]
for i in range(len(mobile_price_element)):
    
    dict={}
    
    mobile_name=mobile_name_element[i].text
    mobile_price=mobile_price_element[i].text

    dict['mobile_name']= mobile_name
    dict['mobile_price']=mobile_price

    scrape_data.append(dict)
print(scrape_data)

dataFrame=pd.DataFrame.from_dict(scrape_data)
dataFrame.to_csv('sg1_Amazon_Mobile_data.csv')
```

```python
#Multipage webscrape

from bs4 import BeautifulSoup
import requests
from requests import get

import pandas as pd 
import numpy as np

from time import sleep
from random import randint

pages=np.arange(1,10)

Mobile_name=[]
Ratings=[]
Prices=[]

for page in pages:
    url="https://www.flipkart.com/search?q=mobile&page=" +str(page)
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    items=soup.find_all('div',class_="_4ddWXP")
    sleep(randint(2,10))

#print("Items divs in page:",page, "in",len(items))

    for item in items:
        name=item.find('a',class_="s1Q9rs").text
        Mobile_name.append(name)
        
        ratings=item.find('div',class_="gUuXy- _2D5lwg").text
        Ratings.append(ratings)
        
        price=item.find('div',class_="_30jeq3").text
        Prices.append(price)
#print(Mobile_name,Ratings,Prices)

dataFrame=pd.DataFrame(
{
    "Mobile Name": Mobile_name,
    "Ratings": Ratings,
    "Prices":Prices
})

print(dataFrame)
#         print(name,ratings,price)

dataFrame.to_csv('Flipkart_mobile_data.csv',index=False)

```
